```python
hr_data_uc = data[['Age','DailyRate','DistanceFromHome', 
                       'EnvironmentSatisfaction', 'HourlyRate',                     
                       'JobInvolvement', 'JobLevel',
                       'JobSatisfaction', 
                       'RelationshipSatisfaction', 
                       'StockOptionLevel',
                        'TrainingTimesLastYear']].copy()
hr_data_uc.head()
```

```python
hr_data_uc.skew()
```

```python
#treating using log
import numpy as np
for col in hr_data_uc.columns:
    if hr_data_uc.skew().loc[col]>0.55:
        hr_data_uc[col]=np.log1p(hr_data_uc[col])
```

```python
#Copy categorical data
hr_data_cat = data[['Attrition', 'BusinessTravel','Department',
                       'EducationField','Gender','JobRole',
                       'MaritalStatus',
                       'Over18', 'OverTime']].copy()
hr_data_cat.head()
```

```python
import sklearn
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import pandas as pd
le=LabelEncoder()
hr_data_cat['Attrition']=le.fit_transform(hr_data_cat['Attrition'])
hr_data_cat.head()
```

```python
hr_data_cat = pd.get_dummies(hr_data_cat)
hr_data_cat.head()
hr_data_cat.shape
```

```python
hr_data_final = pd.concat([hr_data_uc, hr_data_cat], axis=1)
hr_data_final.head()
hr_data_final.shape
```

```python
hr_data_final.shape
```

```python
df_x= hr_data_final.drop(columns=["Attrition"])
df_x.head()
```

```python
y=hr_data_final['Attrition']
y.head()
```

```python
#sacaling down using standardscaler
from sklearn.preprocessing import MinMaxScaler
MS = MinMaxScaler(feature_range=(0, 5))
x=MS.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
x
```
